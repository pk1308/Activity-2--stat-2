```python
import matplotlib.pyplot as plt
import numpy as np

# Define the uniform distribution parameters
a = -3
b = 2

# Create an array of x-values
x = np.linspace(a, b, 100)

# Calculate the CDF of the uniform distribution
cdf_uniform = (x - a) / (b - a)

# Define the absolute value function
def abs_func(x):
  return np.abs(x)

# Calculate the CDF of the absolute value distribution
y = abs_func(x)
cdf_abs = np.where(x >= 0, x, -x) / (b - a)

# Plot the CDFs
plt.plot(x, cdf_uniform, label='Uniform CDF')
plt.plot(x, cdf_abs, label='Absolute Value CDF')

# Set labels and title
plt.xlabel('x')
plt.ylabel('CDF(x)')
plt.title('CDF of Uniform and Absolute Value Distributions')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of x-axis
x = np.linspace(-3, 2, 100)

# Probability density function (pdf) of uniform distribution
def uniform_pdf(x):
  return np.where((x >= -3) & (x <= 2), 1 / 5, 0)

# Probability density function (pdf) of y = |x|
def abs_pdf(x):
  return np.piecewise(x, [x < 0, x >= 0], [lambda x: -1 / 5, lambda x: 1 / 5])

# Plot the pdfs
plt.plot(x, uniform_pdf(x), label='Uniform PDF')
plt.plot(x, abs_pdf(x), label='|x| PDF')

# Set labels and title
plt.xlabel('x')
plt.ylabel('PDF(x)')
plt.title('PDF of Uniform and |x| Distributions')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np

def pdf_X(x):
  """Probability density function of X."""
  if 0 < x < 1:
    return 3 * x**2
  else:
    return 0

def pdf_Y(y):
  """Probability density function of Y = X^2."""
  if y >= 0:  # Consider non-negative y due to X^2
    # Transformation (consider the domain of X)
    x = np.sqrt(y)
    if 0 < x < 1:  # Valid range for f_X
      return 6 * np.sqrt(y) * pdf_X(x)  # Apply transformation formula
    else:
      return 0  # f_X(x) = 0 outside 0 < x < 1
  else:
    return 0  # y must be non-negative

# Sample y values for visualization (optional)
y_vals = np.linspace(0, 1.5, 100)

# Calculate PDF values for the sample y
pdf_y_vals = np.vectorize(pdf_Y)(y_vals)

# Print or plot the PDF (optional)
print("PDF of Y (f_Y(y)) for sample y values:")
for y, pdf_y in zip(y_vals, pdf_y_vals):
  print(f"y = {y:.2f}, f_Y(y) = {pdf_y:.4f}")

# Note: The actual PDF of Y is zero for all y due to the transformation.
```
