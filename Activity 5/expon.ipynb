```python
import pandas as pd
import matplotlib.pyplot as plt 
import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv("satandgpa.csv")
```

```python
df.columns
```

```python
data = df['sat_sum']
```

```python
# Define histogram trace with 10 bins
hist = go.Histogram(
    x=data,
    xbins=dict(start=min(data), end=max(data), size=10),  # Set 10 bins
    marker=dict(color="lightblue")  # Set bar color
)

# Create layout and add trace
fig = go.Figure(data=[hist])
fig.update_layout(title="Histogram of SAT Sum")  # Update title

# Display histogram
fig.show()
```
